```python
import os

def total_files(folder_path):
    num_files = len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
    return num_files

rose_black = "disease/Rose/BlackSpot"
rose_downy = "disease/Rose/DownyMildew"
rose_fresh = "disease/Rose/FreshLeaf"

sunflower_healthy = "disease/Sunflower/FreshLeaf"
sunflower_gray = "disease/Sunflower/Graymold"
sunflower_scar = "disease/Sunflower/Leafscars"


print("Number of healthy leaf images in rose dataset", total_files(rose_fresh))
print("Number of rose downy leaf images in  rose dataset", total_files(rose_downy))
print("Number of rose black  leaf images in rose dataset", total_files(rose_black ))

print("=========================================================")

print("Number of healthy leaf images in sunflower dataset", total_files(sunflower_healthy))
print("Number of gray mold leaf images in sunflower dataset", total_files(sunflower_gray))
print("Number of leaf scars images in sunflower dataset", total_files(sunflower_scar))

```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    rotation_range=30,  # New
    width_shift_range=0.2,  # New
    height_shift_range=0.2,  # New
    brightness_range=[0.8, 1.2],  # New
    validation_split=0.2  # Keep validation split
)

test_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
del train_generator, validation_generator  # Delete old generators

train_generator = train_datagen.flow_from_directory("disease", target_size=(225, 225), batch_size=32, class_mode='categorical', subset='training')
validation_generator = test_datagen.flow_from_directory("disease", target_size=(225, 225), batch_size=32, class_mode='categorical', subset='validation')
```

```python
from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from keras.layers import Dropout

model = Sequential([
    Input(shape=(225, 225, 3)),  
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),  # New Dropout

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),  # New Dropout

    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),  # New Dropout
    Dense(5, activation='softmax')  
])

```

```python

from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

def preprocess_image(image_path, target_size=(225, 225)):
    img = load_img(image_path, target_size=target_size)
    x = img_to_array(img)
    x = x.astype('float32') / 255.
    x = np.expand_dims(x, axis=0)
    return x

x = preprocess_image('disease/Rose/DownyMildew/Downy Mildew  (109).jpg')
```
